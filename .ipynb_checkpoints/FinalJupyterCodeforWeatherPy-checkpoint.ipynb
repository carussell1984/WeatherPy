```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)


```

## Generate Cities List

```python
# List for holding lat_lngs and cities for the first set which is covering the land mass south of brazil (south america) of to the equater
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=5000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=5000)
lat_lngs = zip(lats, lngs)

print(len(lats))
print(len(lngs))
print(lngs)
print(lats)
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs :
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
```

```python
plt.scatter(x=lngs, y=lats, marker="o")
plt.title("Variability in Random Latitude, Longitude Generator")
plt.ylabel("Latitude")
plt.xlabel("Longitude")
#plt.xlim(-100, 100)
plt.yticks(np.arange(-90, 120, 30))
plt.xticks(np.arange(-180, 210, 30))

plt.savefig("Images/Step_1-Random_Lat_Lng_Generator.png")
plt.show()
```

```python
cities_2 = []

for city in cities:
    if " " in city:
        city2 = city.replace(" ", "+")
        if city2 not in cities_2:
            cities_2.append(city2)
    else:
        if city not in cities_2:
            cities_2.append(city)

print(len(cities_2))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "&units=imperial"

#create dictionary lists
city_name = []
temperature = []
latitude = []
humidity_perc = []
cloudiness_perc = []
wind_speed = []
current_time = []
timezone = []
longitude = []

query_url = (url + "appid=" + api_key + units + "&q=")

count = 1

print("------------------------------------------")
print("Begin Processing API Open Weather Requests")
print("------------------------------------------")



for city in cities_2:
    
    
    print(f"Record {count} is processing | {city}")
           
    try:     
        city_weather_requests = requests.get(query_url+city).json()
        city_name.append(city)
        if city_weather_requests["main"]["temp"] is not None:
            temperature.append(city_weather_requests["main"]["temp"])
        if city_weather_requests["coord"]["lat"] is not None:
            latitude.append(city_weather_requests["coord"]["lat"])
        if city_weather_requests["main"]["humidity"] is not None:
            humidity_perc.append(city_weather_requests["main"]["humidity"])
        if city_weather_requests["clouds"]["all"] is not None:
            cloudiness_perc.append(city_weather_requests["clouds"]["all"])
        if city_weather_requests["wind"]["speed"] is not None:
            wind_speed.append(city_weather_requests["wind"]["speed"])
        if city_weather_requests["coord"]["lon"] is not None:
            longitude.append(city_weather_requests["coord"]["lon"])
        if city_weather_requests["timezone"] is not None:
            timezone.append(city_weather_requests["timezone"])
        if city_weather_requests["dt"] is not None:
            current_time.append(city_weather_requests["dt"])
    except: 
        print("city is missing...")
        temperature.append("none")
        latitude.append("none")
        humidity_perc.append("none")
        cloudiness_perc.append("none")
        wind_speed.append("none")
        longitude.append("none")
        timezone.append("none")
        current_time.append("none")
   
    count = count + 1
    time.sleep(1)
    
print("-------------------------------------")
print("End of Weather API Request")
print("-------------------------------------")          
  
            
```

```python
#ensure the lengths are the same - a sanity check
print(len(city_name))
print(len(temperature))
print(len(latitude))
print(len(humidity_perc))
print(len(cloudiness_perc))
print(len(wind_speed))
print(len(current_time))
print(len(timezone))
print(len(longitude))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#savefiles:
    # i the output of two cities file, from the originally generated file and the list with the replaced "+" character
    
import csv

with open("output_data/original_cities.csv", "w", newline = "") as citiesoriginal:
    writer_1 = csv.writer(citiesoriginal)
    for city in cities:    
        writer_1.writerow([city])
    
```

```python
 
with open("output_data/fixed_cities.csv", "w", newline = "") as citiesfixed:
    writer_2 = csv.writer(citiesfixed)
    for city_2 in cities_2:
        writer_2.writerow([city_2])
```

```python
#create dictionary with all the data and then store into data frame
weather_dict = {
    "city": city_name,
    "latitude": latitude,
    "longitude": longitude,
    "time dt": current_time,
    "timezone": timezone,
    "temperature": temperature,
    "humidity_perc": humidity_perc,
    "cloudiness_perc": cloudiness_perc,
    "wind_speed": wind_speed,
}
    

weather_data = pd.DataFrame(weather_dict)
#check to ensure this worked, and then save the data frame as a csv to work with later to plot graphs
weather_data.head()
```
